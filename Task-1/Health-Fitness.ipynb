```python

#import 

import pandas as pd
import matplotlib.pyplot as plt
from wordcloud import WordCloud
import nltk
from nltk.corpus import stopwords
import re



```

```python
# Download stopwords
nltk.download('stopwords')
stop_words = set(stopwords.words('english'))
```

```python
df = pd.read_csv(r"C:\Users\sudhe\OneDrive\Internsip NullClass\Data-Analyst-INTERNSHIP-2025\health_fitness_apps.csv")
```

```python
#Function to clean text
def clean_text(text):
    if pd.isna(text):
        return ""
    text = re.sub(r'\W+', ' ', text.lower())  # Remove special characters & lowercase
    words = text.split()
    return ' '.join([word for word in words if word not in stop_words])
```

```python
#filter 5 star review
df_filtered = df[df['Rating'] == 5]
```

```python
#merge all reviews into one text
text_data = ' '.join(df_filtered['App'].astype(str).apply(clean_text))

# Generate Word Cloud
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(text_data)
wordcloud.to_file("wordgenerator.png")
```

```python
plt.figure(figsize=(10,5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title("Most Frequent Keywords in 5-Star Health & Fitness Reviews")
plt.show()
```
